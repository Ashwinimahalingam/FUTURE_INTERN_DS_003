```python
import pandas as pd
from nltk.corpus import stopwords
import nltk

# Make sure stopwords are downloaded
nltk.download('stopwords')

def clean_text(text):
    if pd.isnull(text):
        return ""
    text = str(text).lower()
    # Keep only letters and spaces
    text = "".join(char for char in text if char.isalpha() or char.isspace())
    stop_words = set(stopwords.words('english'))
    text = " ".join(word for word in text.split() if word not in stop_words)
    return text

feedback_column = "Feedback"   # make sure this column exists in your dataset

df_clean['Cleaned_Feedback'] = df_clean[feedback_column].apply(clean_text)
```

```python
df_clean['Cleaned_Feedback']
```

```python
from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df_clean['Sentiment'] = df_clean['Cleaned_Feedback'].apply(get_sentiment)
df_clean['Polarity'] = df_clean['Cleaned_Feedback'].apply(lambda x: TextBlob(x).sentiment.polarity)

print(df_clean[['Cleaned_Feedback', 'Sentiment', 'Polarity']].head())
```

```python
from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    
    if polarity > 0.1:     # positive if polarity > 0.1
        return "Positive"
    elif polarity < -0.1:  # negative if polarity < -0.1
        return "Negative"
    else:                  # between -0.1 and 0.1 = neutral
        return "Neutral"

# Apply sentiment function
df_clean['Sentiment'] = df_clean['Cleaned_Feedback'].apply(get_sentiment)

# Show counts
print(df_clean['Sentiment'].value_counts())
```

```python
import matplotlib.pyplot as plt

# Count sentiments (include 0 for missing)
sentiment_counts = df_clean['Sentiment'].value_counts().reindex(
    ["Positive", "Negative", "Neutral"], 
    fill_value=0
)

plt.figure(figsize=(6,4))
sentiment_counts.plot(
    kind='bar',
    color=['green','red','gray'],
    edgecolor='black'
)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS

def generate_wordcloud(text, title):
    wordcloud = WordCloud(
        width=200,
        height=150,
        background_color='white',
        stopwords=STOPWORDS
    ).generate(text)
    
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(title, fontsize=16)
    plt.show()

# Word Cloud for all feedback
generate_wordcloud(" ".join(df_clean['Cleaned_Feedback']), "Word Cloud - All Feedback")

# Word Clouds by sentiment
generate_wordcloud(" ".join(df_clean[df_clean['Sentiment']=="Positive"]['Cleaned_Feedback']), "Positive Feedback")
generate_wordcloud(" ".join(df_clean[df_clean['Sentiment']=="Negative"]['Cleaned_Feedback']), "Negative Feedback")
generate_wordcloud(" ".join(df_clean[df_clean['Sentiment']=="Neutral"]['Cleaned_Feedback']), "Neutral Feedback")
```
